```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statistics import mean
%matplotlib inline
sns.set()
```

```python
#A:
def my_func():
    counts_list=[]
    for n in range(100):   
        list_my=[]
        count=0 
        for i in range(10):
            y=np.random.binomial(1, 0.5)
            list_my.append(y)
       
        for j in range(9):
            if list_my[j]==1 and list_my[j+1]==0:
                count+=1
        counts_list.append(count)
    return counts_list
#y=my_func()  
#print(my_func())
pl=sns.distplot(my_func(),hist=False)
pl.set(xlabel='Number of HT ')
pl.set(ylabel='Propability of HT')
plt.title('Distribution of <HT> for a 100 time ')
plt.show()
        
```

```python
#B:
def ret_func():
    new_list=[]
    for i in range (1000):
        t=my_func()
        n=np.mean(t)
        new_list.append(n)
    return new_list

#print (type(n))
sns.distplot(ret_func(),hist=False)       
```

```python
#C:
x=[]   
for n in range(100):
    my_list=[]
    count=0
    for i in range(10):
        y=np.random.binomial(1,0.5)
        my_list.append(y)
        
    for j in range(8):
        if my_list[j]==1 and my_list[j+1]==1 and my_list[j+2]==1:
            count+=1 
    x.append(count)        
    
pl=sns.distplot(x)
pl.set(xlabel='Number of HHH ')
pl.set(ylabel='Propability of HHH')
plt.title('Distribution of <HHH> for a 100 time ')
plt.show()
        
#print(type(x))
```
