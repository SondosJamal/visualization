```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import warnings
warnings.filterwarnings('ignore')
```

                                                  old data set

```python
df=pd.read_csv('all_stocks_5yr.csv')
df[0:2]
```

```python
df['pct_change']=df.groupby(['Name']).close.pct_change()
df[0:3]
```

```python
slim_stock=df[['date','Name','pct_change']]
slim_stock[0:3]
```

```python
reshaped=slim_stock.pivot(index='date',columns='Name',values='pct_change')
reshaped[0:2]
```

```python
x=reshaped[['AAPL','AMZN','KO','UAL','GE']]
y=reshaped['GOOG']
x=sm.add_constant(x)
model=sm.OLS(y,x,missing='drop').fit()
model.summary()
```

                                     The new data set

```python
df2=pd.read_csv('^GSPC (1).csv')
df2.head()
```

```python
df2['Name']='SP'
df2['date']=df2[['Date']]
df2.head()
```

```python
df2['pct_change']=df2.groupby(['Name']).Close.pct_change()
```

```python
sl_stock=df2[['date','Name','pct_change']]
sl_stock[0:2]
```

```python
reshap=sl_stock.pivot(index='date',columns='Name',values='pct_change')
reshap[0:2]
```

```python
df3=pd.merge(reshaped,reshap,on='date',how='inner')
df3.head()
```

```python
x=df3[['AAPL','AMZN','KO','UAL','GE','SP']]
y=df3['GOOG']
x=sm.add_constant(x)
model=sm.OLS(y,x,missing='drop').fit()
model.summary()
```

So after adding the new data set we found that Only AMZN company still significiant but the others not.
